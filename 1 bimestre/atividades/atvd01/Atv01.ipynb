Apresente a soma de todos os argumentos inteiros ou reais recebidos

```java
import java.util.Scanner;

double soma = 0;
double x = 1;
	    
Scanner leitor = new Scanner(System.in);
while (x != 0){
    try {
        System.out.print("(digite um numero, 0 para parar).:"); 
        x = leitor.nextDouble(); 
        soma+=x; // soma recebe x
	} catch (Exception e) {
        leitor.nextLine();
        continue; 
        }
    }
System.out.println("A soma e: " + soma);
```
